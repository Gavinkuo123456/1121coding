```python
import numpy as np
import pandas as pd
import csv
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
railroad = pd.read_csv("railroaddata.csv")
df = railroad
```

```python
railroad
```

```python
fig, ax = plt.subplots()
ax.set_title("Tawian railroad travel distance per passenger")
ax.set_xlabel("year")
ax.set_ylabel("Kms")

x = df["year"]
y = df["km per passengers"]

ax.plot(x,y,'bo-')
```

```python
'''
上圖代表乘客平均搭乘的距離
可見疫情爆發的那兩年，乘客極少搭乘長程，多為短程，可能多是通勤
'''
```

```python
fig, ax = plt.subplots()
ax.set_title("Tawian railroad passengers count")
ax.set_xlabel("year")
ax.set_ylabel("passengers count")

x = df["year"]
y = df["passengers"]

ax.plot(x,y,'bo-')
plt.grid()
```

```python
'''
上圖代表台灣鐵路搭乘人數
可見疫情爆發的那兩年，乘客急遽降低，出遊減少、居家辦公嚴重影響了乘客數
'''
```

```python
fig, ax = plt.subplots()
ax.set_title("Tawian railroad passengers count")
ax.set_xlabel("year")
ax.set_ylabel("passengers count")

x = df["year"]
y = df["passenger twr"]
z = df["passengers hsr"]
a = df["passengers mrt"]

ax.plot(x,y,'ro-',label = "Taiwan railroad")
ax.plot(x,z,'bo-',label = "Taiwan Highspeed rail")
ax.plot(x,a,'yo-',label = "Taiwan MRT")
leg = plt.legend(loc='center right')
plt.grid()
```

```python
'''
上圖代表台鐵、高鐵、捷運的個別搭乘人數
捷運作為城市間通勤的交通工具，原本穩定上升運量的捷運在疫情期間，推測因居家辦公影響，通勤人數急遽降低，造成運量下降
而台鐵與高鐵作為長途交通工具，旅客數雖有減少，但是比較不顯著
'''
```
