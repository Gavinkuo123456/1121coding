```python
import numpy as np
import pandas as pd
import csv
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
railroad = pd.read_csv("railroaddata.csv")
df = railroad
```

```python
railroad
```

```python
trace = sns.scatterplot(railroad,x = 'year',y = 'km per passengers')
```

```python
fig, ax = plt.subplots()
ax.set_title("Tawian railroad passengers count")
ax.set_xlabel("year")
ax.set_ylabel("passengers count")

x = df["year"]
y = df["passengers"]

ax.plot(x,y,'bo-')
plt.grid()
```

```python
fig, ax = plt.subplots()
ax.set_title("Tawian railroad passengers count")
ax.set_xlabel("year")
ax.set_ylabel("passengers count")

x = df["year"]
y = df["passenger twr"]
z = df["passengers hsr"]
a = df["passengers mrt"]

ax.plot(x,y,'ro-')
ax.plot(x,z,'bo-')
ax.plot(x,a,'yo-')
plt.grid()
```
