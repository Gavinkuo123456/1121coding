```python
import os 
import folium
from folium import plugins
import rioxarray as rxr
import earthpy as et
import requests
import earthpy.spatial as es
```

```python
url = 'https://opendata.cwb.gov.tw/api/v1/rest/datastore/O-A0001-001?Authorization=CWB-0305539E-FEE4-45F1-829A-5239DA24147D'
data = requests.get(url)   # 取得 JSON 檔案的內容為文字
data_json = data.json()    # 轉換成 JSON 格式
weather = data_json['records']['Station']
```

```python
m = folium.Map(location=[23.97, 120.97], zoom_start=7)
```

```python
for i in weather:
    folium.Marker(
        location=[i['GeoInfo']['Coordinates'][0]['StationLatitude'], i['GeoInfo']['Coordinates'][0]['StationLongitude']],
        popup=i['StationName']+str(i['WeatherElement']['AirTemperature'])+'°C',icon=folium.Icon(color='green',prefix='fa')).add_to(m)
```
