```python
import requests
```

```python
url = 'https://opendata.cwb.gov.tw/api/v1/rest/datastore/O-A0003-001?Authorization=CWB-0305539E-FEE4-45F1-829A-5239DA24147D'
data = requests.get(url)   # 取得 JSON 檔案的內容為文字
data_json = data.json()    # 轉換成 JSON 格式
weather = data_json['records']['location']
weather
```

```python
for i in weather:
    print(i['locationName'],'Temp:',i['weatherElement'][3]['elementValue'],'°C')
```
