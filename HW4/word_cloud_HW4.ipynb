```python
from wordcloud import WordCloud, STOPWORDS
import numpy as np
import matplotlib.pyplot as plt
import jieba
import jieba.analyse
from collections import Counter
```

```python
dictfile = "C:/Users/gavin/Desktop/python/dict.txt"
stopfile = "C:/Users/gavin/Desktop/python/stopwords.txt"
fontpath = "C:/Users/gavin/Desktop/python/msjh.ttc"
txtfile = "C:/Users/gavin/Desktop/python/hw4_word.txt"
```

```python
jieba.set_dictionary(dictfile)
jieba.analyse.set_stop_words(stopfile)
text = open(txtfile,"r",encoding="utf-8").read()
tags = jieba.analyse.extract_tags(text, topK=25)
seg_list = jieba.lcut(text, cut_all=False)
dictionary = Counter(seg_list)

freq = {}
for ele in dictionary:
    if ele in tags:
        freq[ele] = dictionary[ele]
print(freq)
```

```python
wordcloud = WordCloud(background_color="white", contour_width=3, contour_color='steelblue', font_path= fontpath, width=800, height=600).generate_from_frequencies(freq)
plt.figure()
plt.imshow(wordcloud, interpolation="bilinear")
plt.axis("off")
plt.show()
```
