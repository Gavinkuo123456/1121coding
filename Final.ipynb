```python
import os 
import folium
from folium import plugins
import rioxarray as rxr
import earthpy as et
import requests
import earthpy.spatial as es
```

```python
icon = {
        '陰': 'cloud',
        '晴': 'cun',
        '陰有雨': 'cloud-rain',
        '-99':'question',
        '陰有靄':'smog',
        '有雨':'cloud-showers-heavy',
        '有霧':'smog',
        '多雲':'cloud-sun',
        '陰有霾':'smog'
       }
```

```python
url_1 = 'https://opendata.cwb.gov.tw/api/v1/rest/datastore/O-A0001-001?Authorization=CWB-0305539E-FEE4-45F1-829A-5239DA24147D'
data_1 = requests.get(url_1)   # 取得 JSON 檔案的內容為文字
data_json_1 = data_1.json()    # 轉換成 JSON 格式
weather_1 = data_json_1['records']['Station']
url_2 = 'https://opendata.cwb.gov.tw/api/v1/rest/datastore/O-A0003-001?Authorization=CWB-0305539E-FEE4-45F1-829A-5239DA24147D'
data_2 = requests.get(url_2)   # 取得 JSON 檔案的內容為文字
data_json_2 = data_2.json()    # 轉換成 JSON 格式
weather_2 = data_json_2['records']['Station']
```

```python
m = folium.Map(location=[23.97, 120.97], zoom_start=7)
```

```python
for i in weather_1:
    ##if i['GeoInfo']['CountyName'] == '臺北市' or i['GeoInfo']['CountyName'] == '新北市'or i['GeoInfo']['CountyName'] == '桃園市'or i['GeoInfo']['CountyName'] == '宜蘭縣':
        folium.plugins.BoatMarker(location=[i['GeoInfo']['Coordinates'][0]['StationLatitude'], i['GeoInfo']['Coordinates'][0]['StationLongitude']],
                                  popup=None, icon=None, heading=i['WeatherElement']['WindDirection'], wind_heading=i['WeatherElement']['WindDirection'], wind_speed=i['WeatherElement']['WindSpeed']).add_to(m)
        folium.Marker(
            location=[i['GeoInfo']['Coordinates'][0]['StationLatitude'], i['GeoInfo']['Coordinates'][0]['StationLongitude']],
            popup=i['StationName']+str(i['WeatherElement']['AirTemperature'])+'°C,'+str(round(i['WeatherElement']['AirPressure']/33.86,3))+'in Hg',icon=folium.Icon(icon=icon[i['WeatherElement']['Weather']],color='green',prefix='fa')).add_to(m)
        
```
